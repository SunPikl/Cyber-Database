## Description 

This database contains all the vulnerabilities that were publicly discovered and published by the MITRE foundation, which is a reputable public source for cybersecurity information. This database in particular hosts all vulnerabilities found in 2022 and contains all relevant information to these vulnerabilities.  

 

 

 

## ER Diagram 

Assertions: 

A vulnerability must be sourced from a single product. 1:1 

A product can have several vulnerabilities. 0:N 

A vulnerability can have several threat types. 0:N 

A threat type must correspond to at least one vulnerability. 1:N 

A vulnerability can have one threat score. 

A threat score must apply to at least one vulnerability.

## RM Diagram 
![rm_diagram.png](rm_diagram.png)
 

Attributes: 

cve – String – The Common Vulnerabilities and Exposures identifier  

Primary key of Vulnerability entity 

vendor_project – String – The name of the vendor project associated with the vulnerability 

product – String – the name of the product associated with the vulnerability 

Primary key of Source entity  

vulnerability_name – String – the name of the vulnerability 

date_added – Date – the date that the vulnerability was added to the CISA Known Exploited Vulnerabilities catalog   

short_description – String – brief description of the vulnerability 

required_action – String– the action that is required to remediate the vulnerability 

due_date – Date – the date by which the required action must be completed  

notes – String – additional notes  

grp – String – the group associated with the vulnerability 

pub_date – Date – the date that the vulnerability was published  

cvss – Float – The Common Vulnerability Scoring System score associated with the vulnerability  

Primary key of Threat Score entity 

cwe – String – the Common Weakness Enumeration associated with the vulnerability  

Primary key of Type entity 

vector – String – the vector associated with the vulnerability   

complexity – String – the complexity associated with the vulnerability  

severity – String – the severity level associated with the vulnerability
